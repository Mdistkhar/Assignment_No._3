# 2.Bar chart: A bar chart is a simple visualization that shows the frequency of categorical data. To create a bar chart in Python, you can use the plot.bar() method in Pandas.# 

```python
# Create a bar chart of the `salary` column
df1['Salary'].plot.bar()
# Add a title and labels to the chart
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Count')

# Show the chart
plt.show()
```

```python
# Create a bar chart of the `minimum_nights` column
df2['minimum_nights'].plot.bar()
# Add a title and labels to the chart
plt.title('minimum_nights Distribution')
plt.xlabel('minimum_nights')
plt.ylabel('Count')

# Show the chart
plt.show()
```

# 3. Line chart: A line chart is a visualization that shows the relationship between two variables over time. To create a line chart in Python, you can use the plot.line() method in Pandas.

```python
# Create a line chart of the `host_id` column over time
df2['host_id'].plot.line()

# Add a title and labels to the chart
plt.title('host_id Over Time')
plt.xlabel('host_id')
plt.ylabel('minimum_nights')

# Show the chart
plt.show()
```

```python
# Create a line chart of the `salary` column over time
df1['Salary'].plot.line()

# Add a title and labels to the chart
plt.title('Salary Over Time')
plt.xlabel('Year')
plt.ylabel('Salary')

# Show the chart
plt.show()
```

# 4. Pie chart: A pie chart is a visualization that shows the relative size of different categories. To create a pie chart in Python, you can use the plot.pie() method in Pandas.


```python
# Create a pie chart of the `salary` column
df1['GenderID'].plot.pie()

# Add a title and labels to the chart
plt.title('GenderID')
plt.legend()

# Show the chart
plt.show()
```

```python
# Create a pie chart of the `number_of_reviews` column
df2['number_of_reviews'].plot.pie()

# Add a title and labels to the chart
plt.title('number_of_reviews')
plt.legend()

# Show the chart
plt.show()
```

# 5. Histogram: A histogram is a visualization that shows the distribution of data. To create a histogram in Python, you can use the hist() method in Pandas.

```python
# Create a histogram of the `salary` column
df1['Salary'].hist()

# Add a title and labels to the chart
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Count')

# Show the chart
plt.show()
```

```python
# Create a histogram of the `latitude` column
df2['latitude'].hist()

# Add a title and labels to the chart
plt.title('latitude Distribution')
plt.xlabel('latitude')
plt.ylabel('Count')

# Show the chart
plt.show()
```

# 6. Boxplot: A boxplot is a visualization that shows the distribution of data, including the median, quartiles, and outliers. To create a boxplot in Python, you can use the boxplot() method in Pandas.

```python
sns.boxplot(x = 'Sex', data=df1 , y = 'Salary')
```

```python
sns.boxplot(x = 'latitude', data=df2 , y = 'longitude')
```

# 7. Scatter plot: A scatter plot is a visualization that shows the relationship between two variables. To create a scatter plot in Python, you can use the plot.scatter() method in Pandas.

```python
# Create a scatter plot of the `salary` and `age` columns
df1.plot.scatter('Salary', 'Sex')

# Add a title and labels to the chart
plt.title('Salary vs. Sex')
plt.xlabel('Salary')
plt.ylabel('Sex')
```

```python
# Create a scatter plot of the `longitude` and `latitude` columns
df2.plot.scatter('longitude', 'latitude')

# Add a title and labels to the chart
plt.title('longitude vs. latitude')
plt.xlabel('longitude')
plt.ylabel('latitude')
```
